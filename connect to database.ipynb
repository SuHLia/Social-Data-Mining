```python
REPOS_TABLE_NAME = 'repos'
ISSUE_TABLE_NAME = 'issues'
COMMENT_TABLE_NAME = 'comments'
USER_TABLE_NAME = 'users'

INSERT_REPOS_QUERY = f"""
    INSERT INTO {REPOS_TABLE_NAME}(repos_id, repos_name, repos_language, repos_watchers_count, repos_openIssue_count,
        repos_forks_count, branches_count)
    VALUES(%s,%s,%s,%s,%s,%s,%s)
"""

INSERT_ISSUE_QUERY = f"""
    INSERT INTO {ISSUE_TABLE_NAME}(issue_id, issue_title, labels, comments_number, repos_id)
    VALUES(%s,%s,%s,%s,%s)
"""

INSERT_COMMENT_QUERY = f"""
    INSERT INTO {COMMENT_TABLE_NAME}(comment_id, comments, comments_createdtime, issue_id)
    VALUES(%s,%s,%s,%s)
"""

INSERT_USER_QUERY = f"""
    INSERT INTO {USER_TABLE_NAME}(user_id, user_name, user_image_url, user_location, user_public_repos, user_followers, 
        user_following, user_createdtime, user_updatedtime, repos_id)
    VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
"""

cur = conn.cursor()

with open(CSV_FILE, 'r', encoding="utf8") as f:
    reader = csv.reader(f)

    # read and forget about header
    next(reader)

    for row in reader:
        #print("ORIGINAL", row)
        
        REPOS_SELECT = "SELECT * FROM repos WHERE repos_id=%s"

        EXISTING_REPOS_ID = row[0] # get repos_id from csv
        
        
        ISSUES_SELECT = "SELECT * FROM issues WHERE issue_id=%s"

        EXISTING_ISSUE_ID = row[7] # get issue_id from csv
        
        
        COMMENTS_SELECT = "SELECT * FROM comments WHERE comment_id=%s"

        EXISTING_COMMENT_ID = row[11] # get the comment_id from the csv
        
        
        USERS_SELECT = "SELECT * FROM users WHERE user_id=%s"

        EXISTING_USER_ID = row[14] # get the comment_id from the csv

        try:
            cursor = conn.cursor()
            
            # check the whether the id is in the database.
            cursor.execute(REPOS_SELECT, (EXISTING_REPOS_ID,))
            records_REPOS = cursor.fetchall()
            
            cursor.execute(ISSUES_SELECT, (EXISTING_ISSUE_ID,))
            records_ISSUES = cursor.fetchall()
            
            cursor.execute(COMMENTS_SELECT, (EXISTING_COMMENT_ID,))
            records_COMMENTS = cursor.fetchall()
            
            cursor.execute(USERS_SELECT, (EXISTING_USER_ID,))
            records_USERS = cursor.fetchall()
            
            if records_REPOS:
                print('duplicate record')
            else:
                print('no data')
                cur.execute(INSERT_REPOS_QUERY, (row[0], row[1], row[2],row[3],row[4],row[5],row[6]))
                conn.commit() 

            if records_ISSUES:
                print('duplicate record')
            else:
                print('no data')
                cur.execute(INSERT_ISSUE_QUERY, (row[7], row[8], row[9],row[10],row[0]))
                conn.commit() 
                
                
            if records_COMMENTS:
                print('duplicate record')
            else:
                print('no data')
                cur.execute(INSERT_COMMENT_QUERY, (row[11], row[12], row[13],row[7]))
                conn.commit() 
                
            if records_USERS:
                print('duplicate record')
            else:
                print('no data')
                cur.execute(INSERT_USER_QUERY, (row[14], row[15], row[16],row[17], row[18], row[19], 
                row[20],row[21], row[22], row[0]))
                conn.commit()


        except (Exception, psycopg2.Error) as error :
            print ("Error while fetching data from PostgreSQL", error)
```
